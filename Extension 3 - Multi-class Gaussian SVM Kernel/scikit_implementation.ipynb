# Processing Iris Dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = np.hstack((iris.data[:, 0:1], iris.data[:, 3:]))
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
iris_names=['Iris-setosa','Iris-versicolor','Iris-virginica']
for i in range(0,3):
    plt.scatter(X_train[y_train == i, 0],
                X_train[y_train == i, 1],
            marker='o',
            label='class '+ str(i)+ ' '+ iris_names[i])

plt.xlabel('sepal length[cm]')
plt.ylabel('petal width[cm]')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

print("The entire dataset")
plt.show()

# Test data to predict on
for i in range(0, 3):
    plt.scatter(X_test[y_test == i, 0],
                X_test[y_test == i, 1],
                marker='o',
                label='class '+ str(i)+ ' '+ iris_names[i])

plt.xlabel('sepal length[cm]')
plt.ylabel('petal width[cm]')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

print("The test data to predict on (useful for comparing incorrect labels)")
plt.show()
```

# Section 1 - Linear SVM Dual (Iris)

```python
from sklearn.svm import LinearSVC
svm_dual = LinearSVC(dual=True)
svm_dual.fit(X_train, y_train)
y_pred = svm_dual.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

### Plotting classification results
y_pred = svm_dual.predict(X_test)

X_incorrect = []
y_incorrect = []
X_correct = []
y_correct = []
for i in range(X_test.shape[0]):
    if y_test[i] != y_pred[i]:
        X_incorrect.append(X_test[i])
        y_incorrect.append(y_pred[i])
    else:
        X_correct.append(X_test[i])
        y_correct.append(y_test[i])
X_incorrect = np.array(X_incorrect)
y_incorrect = np.array(y_incorrect)
X_correct = np.array(X_correct)
y_correct = np.array(y_correct)

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
wine_names=['class_0', 'class_1', 'class_2']
for i in range(3):
    plt.scatter(X_correct[y_correct == i, 0],
                X_correct[y_correct == i, 1],
                s=20,
                marker='o',
                color = colors[i],
                label=wine_names[i])
    
mislabeled_names=['class_0 mislabeled', 'class_1 mislabeled', 'class_2 mislabeled']
for i in range(3):
    plt.scatter(X_incorrect[y_incorrect == i, 0],
                X_incorrect[y_incorrect == i, 1],
                s=50,
                marker='x',
                color = colors[i],
                label=mislabeled_names[i])

plt.xlabel('alcohol')
plt.ylabel('proanthocyanins')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.title("Predicted Results")
plt.show()
```

# Processing Wine Dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

wine = load_wine()
X = np.hstack((wine.data[:, 0:1], wine.data[:, 10:11]))
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

wine_names=['class_0','class_1', 'class_2']
for i in range(0, 3):
    plt.scatter(X_train[y_train == i, 0],
                X_train[y_train == i, 1],
                marker='o',
                label=wine_names[i])

plt.xlabel('alcohol')
plt.ylabel('hue')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()

# Test data to predict on
for i in range(0, 3):
    plt.scatter(X_test[y_test == i, 0],
                X_test[y_test == i, 1],
                marker='o',
                label=wine_names[i])

plt.xlabel('alcohol')
plt.ylabel('hue')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

print("The test data to predict on (useful for comparing incorrect labels)")
plt.show()
```
